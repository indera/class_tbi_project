```python
import numpy as np
import pandas as pd

# pd.set_option('display.height', 1000)
# pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 700)
pd.set_option('display.width', 3000)
pd.set_option('display.max_colwidth', 200)

df_clin = pd.read_csv('/Users/asura/git/class_tbi_project/data/clin/COADREAD.clin.merged.txt',
                   sep="\t")
# df_clin.head(50)
rows, cols = df_clin.shape
print("The clinical file dimensions: {} by {}".format(rows, cols))

keep_cols = [
    'patient.patient_id',
    'status', 'status_src', # extra columns for debugging
    'patient.stage_event.pathologic_stage',
    'admin.disease_code',
    'patient.race_list.race',
    'patient.gender',

    'patient.ethnicity',
    'patient.height', 
    'patient.weight',
    'patient.age_at_initial_pathologic_diagnosis',
    'patient.tissue_source_site',

    'patient.tumor_tissue_site',
    'patient.vital_status',
    # 'patient.drugs.drug-2.drug_name',
    # 'patient.drugs.drug-2.therapy_ongoing',
    # 'patient.drugs.drug-3.drug_name',
    # 'patient.drugs.drug-3.therapy_ongoing',
    'patient.colon_polyps_present',

    ]
    
# df_clin_keep = df_clin.loc[df_clin['admin.batch_number'].isin(keep_rows)]
df_clin_keep = df_clin[(df_clin['admin.batch_number'].str.contains("bcr_aliquot_barcode")) 
                           | df_clin['admin.batch_number'].isin(keep_rows)] 

df_clin_keep.head(10)
```

```python
df = df_clin_keep.transpose()
rows, cols = df.shape
print("\n Transposed clinical data: {} patients with {} variables each".format(rows, cols))
df.reset_index(inplace=True)
new_header = df.iloc[0]  # grab the first row for the header
df = df[1:]  # take the data less the header row
df.rename(columns=new_header, inplace=True) # set the header row as the df header

# df = df[keep_rows]
df.to_csv("clin.csv")
df.head(10)

```

```python
def group_stages(x):
    if pd.isnull(x):
        return 'stage_unknown'
    
    if x in ['stage i', 'stage ia', 'stage ib', 'stage ic']:
        return 'stage_1'
    
    if x in ['stage ii', 'stage iia', 'stage iib', 'stage iic']:
        return 'stage_2'
    
        
    if x in ['stage iii', 'stage iiia', 'stage iiib', 'stage iiic']:
        return 'stage_3'
    
            
    if x in ['stage iv', 'stage iva', 'stage ivb', 'stage ivc']:
        return 'stage_4'
    
    
    return None

print("\n Distinct values in column 'patient.stage_event.pathologic_stage': \n{}"
      .format(df['patient.stage_event.pathologic_stage'].unique()))

df['stage'] = df['patient.stage_event.pathologic_stage'].apply(lambda x: group_stages(x))

def print_stats(df):
    group_stage = df[['stage']].groupby('stage').size()
    group_type = df[['admin.disease_code']].groupby('admin.disease_code').size()
    group_race = df[['patient.race_list.race']].groupby('patient.race_list.race').size()
    group_sex = df[['patient.gender']].groupby('patient.gender').size()
    group_ethnicity = df[['patient.ethnicity']].groupby('patient.ethnicity').size()
 
    group_vital_status = df[['patient.vital_status']].groupby('patient.vital_status').size()
    group_polyps = df[['patient.colon_polyps_present']].groupby('patient.colon_polyps_present').size()

    ht_mean = df['patient.height'].astype(float).mean()
    ht_std = df['patient.height'].astype(float).std()
    wt_mean = df['patient.weight'].astype(float).mean()
    wt_std = df['patient.weight'].astype(float).std()
    age_mean = df['patient.age_at_initial_pathologic_diagnosis'].astype(float).mean()
    age_std = df['patient.age_at_initial_pathologic_diagnosis'].astype(float).std()

    print("\n === Printing stats for frame with {} rows".format(len(df)))
    print("\n > Patient counts by cancer stage: \n{}".format(group_stage))
    print("\n > Patient counts by cancer type: \n{}".format(group_type))
    print("\n > Patient counts by race: \n{}".format(group_race))
    print("\n > Patient counts by sex: \n{}".format(group_sex))
    print("\n > Patient counts by ethnicity: \n{}".format(group_ethnicity))
    print("\n > Patient counts by vital_status: \n{}".format(group_vital_status))
    print("\n > Patient counts by polyps_present: \n{}".format(group_polyps))
    
    print("\n > Height mean: {} (std: {})".format(ht_mean, ht_std))
    print("\n > Weight mean: {} (std: {})".format(wt_mean, wt_std))
    print("\n > Age mean: {} (std: {})".format(age_mean, age_std))
    
print_stats(df)
```

```python
df_expr = pd.read_csv('/Users/asura/git/class_tbi_project/data/expr/COADREAD.rnaseqv2__illuminahiseq_rnaseqv2__unc_edu__Level_3__RSEM_genes_normalized__data.data.txt',
                   sep="\t", nrows=1)


# The first row in the expression data contain the aliquot ids (except the first column)
filter_values = sorted(df_expr.columns[1:]) # 433 values
filter_values = [x.lower() for x in filter_values]
df_filter = pd.DataFrame(filter_values)
# df_filter[0] = df_filter[0].str.lower()


df_filter.rename(columns={0: "status"}, inplace=True)
df_filter.head()


def check_pat_data(x, values):
    for val in values:
        if val in x.values:
            return val
    return None


def check_pat_data_src(x, values):
    """Find the column which gave a match"""
    for val in values:
        if val in x.values:
            return list(x.keys())[list(x.values).index(val)]
    return None

df['status'] = df.apply(lambda x: check_pat_data(x, filter_values), axis=1)
df['status_src'] = df.apply(lambda x: check_pat_data_src(x, filter_values), axis=1)


null_index = df['status'].isnull()
matched = df[~null_index]
unmatched = df[null_index]

print("Unmatches patients: {}".format(len(unmatched)))
print("Matched patients: {} using: \n{}".format(len(matched), df[['status_src']].groupby('status_src').size()))
matched[keep_cols].head()
```
